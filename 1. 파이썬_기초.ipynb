```python
# for 는 반복문 => 똑같은 작업을 할 수 있도록 횟수를 지정하는 반복문
# 입력하는 단의 구구단 
# 예) 3 입력 => 3 * 1 = 3 ........... 3 * 9 = 27

n = int(input('숫자를 입력하세요>>>')) # 내가 원하는 숫자 출력
for i in range(1, 10):
  print(f'{n} * {i} = {n*i}') # {} 는 안에 있는 변수의 값을 출력
# print(n, '*', i, '=', n * i ) 도 가능
```

```python
# 나머지 연산자의 중요성

# 0. 사용자 입력을 숫자로 받기 
# 1. 1부터 입력수 까지 출력 => for 문 필요
# 2. 3, 6, 9 인지 판단하여 X 로 변환
n = int(input('숫자를 입력해주세요>>> '))

for i in range(1, n+1):
# 3의 배수면 3으로 나누면 나머지가 0
  if i % 3 == 0: # i % 3 == 0: => i 나누기 3 의 나머지가 0 일때 => i 가 3 의 배수일 때 
    print('X', end = ' ')
  else:
    print(i, end = ' ')
```

```python
# 내가 생각한 숫자 맞추기
# 10번의 기회동안 숫자를 맞춤
# 10번의 기회 => 10번동안 입력을 받음
# 10번 동안 => 10번 반복 => for 문 필요
# 입력받음 => input() 필요
# 내가 입력한 숫자가 맞는지 검사 => if 문 필요
# 10번 반복할 동안 검사를 매번 수행 for 문 안에서 입력, 검사 받기



answer = 15

for i in range(10):
  print("1~100 까지의 숫자 중, 내가 생각한 숫자는?")
  num = int(input(f'{i+1}번째 기회 >>> '))
  if num == answer:
      print('정답입니다!')
      break
  elif num < answer:
    print('UP!')
  elif num > answer:
    print('Down!')
```

```python
# 정수 1개를 입력받아 1부터 그 수까지 짝수의 합을 구해보자
# 1 부터 n 까지 반복 => for문
# 반복하면서 나오는 숫자가 짝수 인지 검사 => % 2 == 0
# 합 구하기 => 합을 저장하는 변수 만들기
# 변수를 누적해서 저장하기

num = int(input("아무 숫자나 입력해주세요>>> "))

sum = 0
for i in range(1, num + 1):
  # i 에 1 ~ n 까지의 값이 들어옴
  # i 가 짝수인지 판단
  if i % 2 == 0:
    # sum 에다가 i 를 누적 => sum 에 현재 i 값을 보관
    sum = sum + i

print("1부터 입력한 숫자까지의 짝수의 합계는", sum, "입니다.")
```

```python
# 정수 1개를 입력받아 1부터 그 수까지 홀수의 합을 구해보자
# 1 부터 n 까지 반복 => for문
# 반복하면서 나오는 숫자가 홀수 인지 검사 => % 2 != 0
# 합 구하기 => 합을 저장하는 변수 만들기
# 변수를 누적해서 저장하기


num = int(input("아무 숫자나 입력해주세요>>> "))

sum = 0
for i in range(1, num + 1):
  if i % 2 != 0:
    sum = sum + i

print("1부터 입력한 숫자까지의 홀수의 합계는", sum, "입니다.")
```

```python
#리스트에서 index를 쓰지않고 해당 위치 반환

a = [1,2,3,4,5,6,7]  #크기고정
# a에서 쓸수 있는 기능
# pos = a.index(4) => 4의 위치는?
# print(pos) => 3
size = len(a) # a 배열의 길이     /    a.length 는 자바
find = int(input())

#for문을 이용 => a의 길이(크기) 만큼 반복 검사

for i in range(size):
  # a[i] 번지가 find 와 같은가?
  if a[i] == find:
    print(i)
    break

###결과###
# 4입력 =>   3(번지)출력
```

```python
# 마지막부터 값 찾기
# hint = range 활용

a = [1,2,3,4,5,6,7,8,9]  #크기고정
size = len(a) 
find = int(input())
# range(5) 0~4까지
# range(1, 5) 1~4까지
# range(1, 10, 2) 1~9까지 2씩 증가
# 10부터 0까지 1씩 감소 => range(11, 0, -1)
for i in range(size, -1, -1):
  if a[i - 1] == (find - 1):
    print(i)
    break

# 더 간편한 방법
a = [1,2,3,4,5,6,7,8,9]  #크기고정
size = len(a) 
find = int(input())

for j in range((size - 1), -1, -1):
  if a[j] == find:
    print(j)
    break
```

```python
# for i in range(size):   로 마지막부터 값 찾기

a = [1,2,3,4,5,6,7,8,9]  #크기고정
size = len(a) 
find = int(input())

for i in range(size): 
  # 0 번지 => 9만들기
  pos = (size - 1 - i) # i = 0 일때, 10 - 1 - i => 9 - 0 = 9번지 찾기
  if a[pos] == find:
    print(pos)
    break
```
